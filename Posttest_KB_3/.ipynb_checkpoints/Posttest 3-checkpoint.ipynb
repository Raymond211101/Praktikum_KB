## Visualisasi Data

### Import Library

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### Membaca dataset

```python
df = pd.read_csv("Dataset/data.csv")
```

```python
df1 = df.iloc[:2000] ### Membatasi Record menjadi 2000
```

### Pie Chart

```python
dfTahun = df1[df["year"] >= 2015]
data_pie = dfTahun["year"].value_counts().rename_axis("year").reset_index(name="Id_count")
```

```python
plt.figure(figsize=(15,15))
plt.pie(data_pie.Id_count, labels=data_pie.year,startangle=90,autopct="%.1f%%")
plt.title("Perbandingan Persentase Tahun")
plt.show()
```

#### Kesimpulan Dari Visualisasi Pie Chart
Dapat dilihat di atas bahwa persentase terbanyak berada pada tahun 2020 yang artinya pada tahun tersebut manga, manhua dan manhwa paling banyak terbit pada tahun tersebut sedangkan pada tahun 2022 merupakan tahun dimana manga,manhua dan manhwa paling sedikit terbit

### Grafik Boxplot

```python
xlabel = dfTahun["year"]
ylabel = df1["rating"]
sns.boxplot( x= xlabel,y=ylabel)
plt.title("Persebaran Data Rating dari Tahun ke Tahun")
plt.show()
```

#### Kesimpulan dari Visualisasi Boxplot
Dilihat dari grafik boxplot diatas dapat disimpulkan bahwa dari tahun ke tahun persebaran datanya kurang lebih sama yaitu sekitar 4,40 sampai 4,50 begitu juga dengan nilai maksimum dan outlinernya yaitu pada 4,60 dan 4,70 kecuali pada tahun 2022 tidak memiliki outliner.

### Grafik Garis 

```python
dfGroupby = dfTahun.groupby("year")["rating"].mean()
sns.lineplot(data = dfGroupby,x=dfGroupby.index,y=dfGroupby)
plt.title("Grafik Rating dari Tahun ke Tahun")
plt.show()
```

#### Kesimpulan dari Visualisasi Plot dan Garis
Dilihat dari grafik di atas dapat disimpulkan bahwa awalnya rating mengalami penurunan dari tahun 2015 sampai 2016 kemudian mengalami kenaikan pada tahun 2017, lalu mengalami penurunan yang drastis pada tahun 2018 dan naik turun sampai pada tahun 2021, kemudian pada tahun 2022 mengalami kenaikan lagi
