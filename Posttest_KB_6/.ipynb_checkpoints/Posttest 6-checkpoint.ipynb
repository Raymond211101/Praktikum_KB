## Unsupervised Learning

### Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

### Read Data

```python
df = pd.read_csv('Dataset/Customer Segmentation.csv')
```

### Encoding Data

```python
encoder = OrdinalEncoder()
```

```python
x_encoder = df.drop(["Segmentation",],axis=1)
```

```python
x_encoder = pd.get_dummies(x_encoder)
```

```python
x_encoder.head(5)
```

### Metode K-Means

#### Menentukan jumlah cluster

```python
model = KMeans(n_clusters=3) 
```

#### Memilih atribut yang akan dilatih

```python
points = x_encoder[["Age","Work_Experience"]]
```

```python
model.fit(points)
```

```python
labels = model.labels_
```

```python
labels
```

### Evaluasi Hasil Pemodelan

#### Membuat dataframe dengan label dan segmentation sebagai kolom

```python
df_KMeans = pd.DataFrame({'labels' : labels, 'Segmentation':df.Segmentation})
crosstab = pd.crosstab(df_KMeans['labels'],df_KMeans['Segmentation'])
```

```python
crosstab
```

### Melihat Centroid Dari Model Yang Telah Dibuat

```python
x = points["Age"]
y = points["Work_Experience"]
```

```python
centroids = model.cluster_centers_
centroids_X = centroids[:,0]
centroids_Y = centroids[:,1]
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x,y,alpha=0.5, c=labels)
plt.scatter(centroids_X,centroids_Y,marker='D',s=50,label='centroid')
plt.xlabel('Age')
plt.ylabel('Work Experience')
plt.legend()
plt.show()
```

### Menentukan Jumlah Cluster

```python
kluster_range = range(1, 9)
inertias = []

for k in kluster_range:
    model = KMeans(n_clusters=k)
    model.fit(points)
    inertias.append(model.inertia_)
```

```python
plt.figure(figsize=(10,10))
plt.plot(kluster_range, inertias, '-o')
plt.xlabel('Jumlah Cluster')
plt.ylabel('inertia')
plt.xticks(kluster_range)
plt.show()
```
